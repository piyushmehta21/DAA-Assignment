```
import time
import matplotlib.pyplot as plt
import sys

# Recursive Fibonacci function
def rec_fibo(num):
    if num <= 1:
        return num
    else:
        return rec_fibo(num - 1) + rec_fibo(num - 2)

# Range of n values to test
inputs = list(range(1, 28))
time_taken = []
memory_usage = []

# Calculate time & space usage
for val in inputs:
    t1 = time.time()
    rec_fibo(val)
    t2 = time.time()

    time_taken.append(t2 - t1)

    mem_per_frame = sys.getsizeof(val)
    memory_usage.append(val * mem_per_frame)

# Plotting results
plt.figure(figsize=(9, 5))
plt.plot(inputs, time_taken, marker='s', color='red', label='Execution Time (s)')
plt.plot(inputs, memory_usage, marker='^', color='purple', label='Memory (bytes)')
plt.title('Recursive Fibonacci: Time vs Space Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Measured Value')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Print complexity analysis
print("\n=== Recursive Fibonacci Performance ===")
print("Function Behavior:")
print(" - Input: integer n (position in Fibonacci sequence)")
print(" - Output: nth Fibonacci number")
print("\nTime Complexity:")
print(" - Base Case: O(1) when n <= 1")
print(" - Worst Case: O(2^n) due to repeated calls")
print(" - Overall: Exponential Growth in runtime")
print("\nSpace Complexity:")
print(" - O(n) from recursive call stack depth")
```

```
import time
import matplotlib.pyplot as plt
import sys

# Fibonacci using Dynamic Programming (Tabulation)
def fib_dp(num):
    if num <= 1:
        return num
    table = [0, 1]
    for i in range(2, num + 1):
        table.append(table[i - 1] + table[i - 2])
    return table[num]

# Different input sizes
inputs = [200, 800, 1500, 3000, 6000, 12000]
time_taken = []
memory_used = []

# Measuring performance
for val in inputs:
    t_start = time.time()
    fib_dp(val)
    t_end = time.time()
    time_taken.append(t_end - t_start)

    bytes_per_val = sys.getsizeof(val)
    memory_used.append(val * bytes_per_val)

# Plot results
plt.figure(figsize=(9, 5))
plt.plot(inputs, time_taken, marker='s', color='blue', label='Execution Time (sec)')
plt.plot(inputs, memory_used, marker='^', color='green', label='Memory Usage (bytes)')
plt.title('Dynamic Programming Fibonacci: Time vs Space Analysis')
plt.xlabel('n (input size)')
plt.ylabel('Performance Metrics')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```
import time
import random
import sys
import matplotlib.pyplot as plt

# Bubble Sort Function
def bubble_sort_algo(lst):
    n = len(lst)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    return lst

# Different input sizes
arr_sizes = [200, 600, 1200, 2500, 3500, 4500, 6000]
exec_time = []
mem_usage = []

# Measure time & space
for length in arr_sizes:
    data = [random.randint(1, 9999) for _ in range(length)]

    mem_usage.append(sys.getsizeof(data))

    t_start = time.time()
    bubble_sort_algo(data.copy())
    t_end = time.time()

    exec_time.append(t_end - t_start)

# Plot results
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(arr_sizes, exec_time, marker='s', color='red', label='Time Taken')
ax1.set_xlabel('Input Size (n)')
ax1.set_ylabel('Execution Time (sec)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(arr_sizes, mem_usage, marker='^', color='green', linestyle='--', label='Memory Usage')
ax2.set_ylabel('Memory (bytes)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.title('Bubble Sort: Time vs Memory Complexity')
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

# Complexity Explanation
print("\n=== Bubble Sort Complexity Analysis ===")
print("Function Input : List of random integers")
print("Function Output: Sorted list in ascending order")

print("\nTime Complexity:")
print(" - Best Case   : O(n) when array is already sorted")
print(" - Average Case: O(n^2)")
print(" - Worst Case  : O(n^2) when array is in descending order")

print("\nSpace Complexity:")
print(" - O(1) since sorting is done in place with minimal extra variables")
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

# Merge Sort Algorithm
def merge_sort_algo(lst):
    if len(lst) <= 1:
        return lst
    mid = len(lst) // 2
    left_half = merge_sort_algo(lst[:mid])
    right_half = merge_sort_algo(lst[mid:])
    return merge_lists(left_half, right_half)

def merge_lists(left, right):
    merged = []
    l = r = 0
    while l < len(left) and r < len(right):
        if left[l] <= right[r]:
            merged.append(left[l])
            l += 1
        else:
            merged.append(right[r])
            r += 1
    merged.extend(left[l:])
    merged.extend(right[r:])
    return merged

# Input sizes for analysis
sizes = [20, 60, 120, 300, 600, 1000]
time_taken = []
memory_used = []

# Performance measurement
for n in sizes:
    arr = [random.randint(1, 2000) for _ in range(n)]

    start_time = time.time()
    merge_sort_algo(arr)
    end_time = time.time()
    time_taken.append(end_time - start_time)

    memory_used.append(n * sys.getsizeof(arr[0]))

# Plotting results
plt.figure(figsize=(9, 5))
plt.plot(sizes, time_taken, marker='s', color='red', label='Execution Time (s)')
plt.plot(sizes, memory_used, marker='^', color='purple', label='Memory Usage (bytes)')
plt.title('Merge Sort: Time vs Space Complexity')
plt.xlabel('Array Size (n)')
plt.ylabel('Measured Values')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

# Complexity Discussion
print("\n=== Merge Sort Complexity Analysis ===")
print("Function Input : Array of integers")
print("Function Output: Sorted array in ascending order")

print("\nTime Complexity:")
print(" - Best Case   : O(n log n)")
print(" - Average Case: O(n log n)")
print(" - Worst Case  : O(n log n)")

print("\nSpace Complexity:")
print(" - O(n), extra memory needed for temporary arrays during merging")
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

# Insertion Sort Function
def insertion_sort(nums):
    for idx in range(1, len(nums)):
        current = nums[idx]
        pos = idx - 1
        while pos >= 0 and nums[pos] > current:
            nums[pos + 1] = nums[pos]
            pos -= 1
        nums[pos + 1] = current
    return nums

# Different input sizes
sizes = [20, 60, 120, 240, 480, 960]
time_results = []
memory_usage = []

for n in sizes:
    data = [random.randint(1, 2000) for _ in range(n)]

    # Measuring time
    start_time = time.time()
    insertion_sort(data.copy())
    end_time = time.time()
    time_results.append(end_time - start_time)

    # Estimating space
    memory_usage.append(sys.getsizeof(data) + (n * sys.getsizeof(data[0])))

# Plotting results
plt.figure(figsize=(9, 5))
plt.plot(sizes, time_results, marker='s', color='red', label='Execution Time (s)')
plt.plot(sizes, memory_usage, marker='^', color='purple', label='Memory Used (bytes)')
plt.title('Insertion Sort: Time vs Space Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Measured Value')
plt.legend()
plt.grid(True, linestyle='--')
plt.show()

# Complexity Analysis
print("\n--- Insertion Sort Overview ---")
print("Input: List of integers")
print("Output: Sorted list in ascending order")

print("\nTime Complexity:")
print("  Best Case: O(n)  already sorted data")
print("  Average Case: O(n^2)")
print("  Worst Case: O(n^2)  reverse sorted input")

print("\nSpace Complexity:")
print("  O(1)  works in-place, minimal extra space")
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

# Quick Sort Function
def quick_sort(data):
    if len(data) <= 1:
        return data
    pivot_index = len(data) // 2
    pivot_value = data[pivot_index]
    smaller = [val for val in data if val < pivot_value]
    equal = [val for val in data if val == pivot_value]
    greater = [val for val in data if val > pivot_value]
    return quick_sort(smaller) + equal + quick_sort(greater)

# Different input sizes
sizes = [15, 60, 150, 300, 500, 900]
time_taken = []
memory_used = []

for n in sizes:
    sample = [random.randint(1, 2000) for _ in range(n)]

    # Execution time
    start = time.time()
    quick_sort(sample)
    end = time.time()
    time_taken.append(end - start)

    # Space usage
    memory_used.append(sys.getsizeof(sample) + (n * sys.getsizeof(sample[0])))

# Plotting results
plt.figure(figsize=(9, 5))
plt.plot(sizes, time_taken, marker='D', color='brown', label='Execution Time (s)')
plt.plot(sizes, memory_used, marker='*', color='teal', label='Memory Usage (bytes)')
plt.title('Quick Sort: Time vs Space Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Measured Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Complexity Analysis
print("\n--- Quick Sort Overview ---")
print("Input  List of integers")
print("Output  Sorted list in ascending order")

print("\nTime Complexity:")
print("  Best Case: O(n log n) when pivot balances partitions")
print("  Average Case: O(n log n)")
print("  Worst Case: O(n^2) when pivot is consistently extreme")

print("\nSpace Complexity:")
print("  Best/Average: O(log n) due to recursion stack")
print("  Worst: O(n) if recursion depth is maximum")
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

# Selection Sort Function
def selection_sort_algo(lst):
    length = len(lst)
    for i in range(length):
        smallest_idx = i
        for j in range(i + 1, length):
            if lst[j] < lst[smallest_idx]:
                smallest_idx = j
        lst[i], lst[smallest_idx] = lst[smallest_idx], lst[i]
    return lst

# Input sizes for testing
sizes = [20, 60, 120, 250, 500, 1000]
time_taken = []
memory_used = []

for n in sizes:
    data = [random.randint(1, 2000) for _ in range(n)]

    # Measure execution time
    start_time = time.time()
    selection_sort_algo(data.copy())
    end_time = time.time()
    time_taken.append(end_time - start_time)

    # Measure space usage
    memory_used.append(sys.getsizeof(data) + n * sys.getsizeof(data[0]))

# Plotting results
plt.figure(figsize=(9, 5))
plt.plot(sizes, time_taken, marker='s', color='green', label='Execution Time (s)')
plt.plot(sizes, memory_used, marker='^', color='orange', label='Memory Usage (bytes)')
plt.title('Selection Sort: Time vs Space Complexity')
plt.xlabel('Input Size (n)')
plt.ylabel('Measured Values')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Complexity Analysis
print("\n=== Selection Sort Analysis ===")
print("Input   List of integers")
print("Output  Sorted list in ascending order")

print("\nTime Complexity:")
print("  Best Case: O(n^2)  no early termination")
print("  Average Case: O(n^2)")
print("  Worst Case: O(n^2)")

print("\nSpace Complexity:")
print("  O(1)  in-place sorting, minimal extra memory used")
```

```
import time
import matplotlib.pyplot as plt
import sys
import random

# Iterative Binary Search Function
def binary_search_iter(arr, key):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid_idx = (left + right) // 2
        if arr[mid_idx] == key:
            return mid_idx
        elif arr[mid_idx] < key:
            left = mid_idx + 1
        else:
            right = mid_idx - 1
    return -1

# Input sizes for analysis
sizes = [20, 60, 120, 250, 500, 1000, 2000]
time_taken = []
memory_used = []

for n in sizes:
    data = sorted([random.randint(1, 20000) for _ in range(n)])
    target_val = random.choice(data)

    # Measure execution time
    start_time = time.time()
    binary_search_iter(data, target_val)
    end_time = time.time()
    time_taken.append(end_time - start_time)

    # Estimate memory usage
    memory_used.append(sys.getsizeof(data) + n * sys.getsizeof(data[0]))

# Plotting results
plt.figure(figsize=(9, 5))
plt.plot(sizes, time_taken, marker='s', color='green', label='Execution Time (s)')
plt.plot(sizes, memory_used, marker='^', color='orange', label='Memory Usage (bytes)')
plt.title('Binary Search: Time vs Space Complexity')
plt.xlabel('Array Size (n)')
plt.ylabel('Measured Values')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Complexity Analysis
print("\n=== Binary Search Overview ===")
print("Input   Sorted list of integers and a target element")
print("Output  Index of target in list, or -1 if absent")

print("\nTime Complexity:")
print("  Best Case   : O(1)  Target at middle on first attempt")
print("  Average Case: O(log n)")
print("  Worst Case  : O(log n)  Target at end or not present")

print("\nSpace Complexity:")
print("  Iterative   : O(1)  Constant extra space")
print("  Recursive   : O(log n)  Recursion stack depth")
```
